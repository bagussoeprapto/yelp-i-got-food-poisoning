```python
# Not every CSV has encoding, yet better safe than sorry
raw_restaurants_df = pd.read_csv("raw_all_restaurants.csv", encoding="ISO-8859-1")
```

```python
raw_restaurants_df = raw_restaurants_df.sort_values(["facility_id","activity_date"], ascending=False)

#raw_restaurants_df.drop_duplicates(subset=["activity_date"], take_last=False)

raw_restaurants_df.head()
```

```python
#NOTE TO TEAM: "Unnamed: 0", "Unnamed: 0.1","index", and "owner_id" will not drop--the latter cannot be found in the axis ... not sure how to drop
raw_restaurants_df.drop(["Unnamed: 0", "Unnamed: 0.1", "index", "owner_id","serial_number", "service_code", "service_description", "employee_id", "owner_name", "program_element_pe", "program_name", "program_status"], axis=1, inplace=True)

raw_restaurants_df.head()
```

```python
raw_restaurants_df = raw_restaurants_df.rename(columns={"unnamed (index)": "Index #",
"activity_date": "Inspection Date",
"facility_name": "Restaurant",
"score": "Score",
"grade": "Grade",
"facility_id": "Facility ID",
"facility_address": "Address",
"facility_city": "City",
"facility_state": "State",
"facility_zip": "Zip", 
"index": "IGNORE",                                  
"owner_id": "Owner ID",
"pe_description": "Type",
"record_id": "Inspection ID",           
"food_type": "Cuisine",
"in_business": "Active Business",
"latitude": "Latitude",
"longitude": "Longitude", 
"price": "Price",
"rating": "Rating",
"review_count": "# of Reviews",
"yelp_id": "Yelp ID"
})

raw_restaurants_df.head()
```

```python
raw_restaurants_df.drop_duplicates(raw_restaurants_df.columns.difference(["Inspection Date"]), keep="first", inplace=True)

raw_restaurants_df.head()
```
