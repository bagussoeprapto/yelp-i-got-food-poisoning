```python
#Dependencies
import os
import pandas as pd
import numpy as np
import glob
from matplotlib import pyplot as plt
```

```python
filepath = "called/**"
all_csv_files = glob.glob(os.path.join(filepath, "*.csv"), recursive = True)

grades_n_yelp_df = (pd.read_csv(f) for f in all_csv_files)
raw_restaurants_df = pd.concat(grades_n_yelp_df, ignore_index=True)


#all_restaurants_df=pd.concat(map(pd.read_csv, filepath))
```

```python
raw_restaurants_df.head()
```

```python
# Output BIG dataframe to CSV
raw_restaurants_df.to_csv("raw_all_restaurants.csv", header=True)
```

```python
unique_restaurants = len(raw_restaurants_df["facility_name"].unique())

print(unique_restaurants)
```

```python
#Do a count of all columns
raw_restaurants_df.count()
```

```python
#Drop unnecessary columns
#NOTE TO TEAM: "Unnamed:" and "owner_id" will not drop--the latter cannot be found in the axis ... not sure how to drop
raw_restaurants_df.drop(["serial_number", "service_code", "service_description", "employee_id", "owner_name", "program_element_pe", "program_name", "program_status"], axis=1).rename(columns={"unnamed (index)": "Index #",
"activity_date": "Inspection Date",
"facility_name": "Restaurant",
"score": "Score",
"grade": "Grade",
"facility_id": "Facility ID",
"facility_address": "Address",
"facility_city": "City",
"facility_state": "State",
"facility_zip": "Zip",  
"owner_id": "Owner ID",
"pe_description": "Type",
"record_id": "Inspection ID",           
"food_type": "Cuisine",
"in_business": "Active Business",
"latitude": "Latitude",
"longitude": "Longitude", 
"price": "Price",
"rating": "Rating",
"review_count": "# of Reviews",
"yelp_id": "Yelp ID"
})
```

```python
#Create groupbys to calculate averages
avg_inspection_score = raw_restaurants_df.groupby(["facility_name"])["score"].mean()
avg_yelp_rating = raw_restaurants_df.groupby(["facility_name"])["rating"].mean()

# Convert groupby elements into a DataFrame
restaurant_averages_df = pd.DataFrame({"Avg. Inspection Score": avg_inspection_score, "Avg. Yelp Rating": avg_yelp_rating})

#Make the data pretty
restaurant_averages_df["Avg. Inspection Score"]=restaurant_averages_df["Avg. Inspection Score"].map("{:,.2f}".format)
restaurant_averages_df["Avg. Yelp Rating"]=restaurant_averages_df["Avg. Yelp Rating"].map("{:,.2f}".format)
                                                               
                                                                                  
# View dataframe
restaurant_averages_df
```
