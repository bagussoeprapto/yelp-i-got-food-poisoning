```python
import sys
import time
import pandas as pd

```

```python
df = pd.read_csv('0la_restaurants.csv')
df.head()
```

```python
total_calls = int((df['program_name'].count())/96)
print(total_calls)
```

```python
for x in range(total_calls):
    i = x + 1
    p = round((i/total_calls)*100,1)
    sys.stdout.write(f"\rCalling {i} of {total_calls} -- {p}% complete")
    #sys.stdout.write("\rCalling " + str(x) + " of " + str(total_calls) + " -- " + str(p) + "% complete")
    time.sleep(1)
    sys.stdout.flush()
```
