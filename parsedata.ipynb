```python
import pandas as pd
```

```python
city = pd.read_csv("health_scores.csv")
city.head()
```

```python
rest_only = city[city['pe_description'].str.contains("RESTAURANT") == True]
rest_only.head()
```

rest_only.value_counts()

```python
max_rows = 2500
dataframes = []
while len(rest_only) > max_rows:
    top = rest_only[:max_rows]
    dataframes.append(top)
    rest_only = rest_only[max_rows:]
else:
    dataframes.append(rest_only)
```

```python
for _, frame in enumerate(dataframes):
    frame.to_csv(str(_)+'la_restaurants.csv', index=False)
```
